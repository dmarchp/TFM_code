```python
import pandas as pd
import matplotlib.pyplot as plt

df0 = pd.read_csv('bots_concentric_radial_dist_N_35_ar_20.0_er_0.0.csv')
df1 = pd.read_csv('bots_concentric_radial_dist_N_35_ar_20.0_er_1.5.csv')
```

```python
fig, ax = plt.subplots()
ax.plot(df0['radius'], df0['N'], label='$r_e = 0.0$')
ax.plot(df1['radius'], df1['N'], label='$r_e = 1.5$')
fig.legend()
```

```python
from numpy import sqrt
df0['stdN'] = sqrt(df0['varN'])
df1['stdN'] = sqrt(df1['varN'])
```

```python
fig, ax = plt.subplots()
ax.plot(df0['N'], df0['stdN'], label='$r_e = 0.0$')
ax.plot(df1['N'], df1['stdN'], label='$r_e = 1.5$')
fig.legend()
```
