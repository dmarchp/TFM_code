```python
ids = pd.unique(df_nested_list['ID'])
Nbots = len(ids)
print(Nbots)
```

```python
from collections import Counter
count = Counter(list(df_nested_list.loc[df_nested_list['ID']==0]['ticks']))
```

```python
# is there any tick with more than 1 config?
ticks_plus1_configs = []
for k,v in count.items():
    if v>1:
        ticks_plus1_configs.append(k)
        
if(ticks_plus1_configs):
    print(f"There are {len(ticks_plus1_configs)} ticks with more than 1 config")
```
