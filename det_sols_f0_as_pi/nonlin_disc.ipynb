```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def potential(f1, l, pi, r):
    a, b = -1/4, 1/3
    c = -(r+pi)/(2*l)
    d = pi/l
    return a*f1**4 + b*f1**3 + c*f1**2 + d*f1

def potential2(f1, pibar, rbar): # pibar = pi/(2l), rbar = r/(2l)
    a, b = -1/4, 1/3
    c = -(pibar+rbar)
    d = 2*pibar
    return a*f1**4 + b*f1**3 + c*f1**2 + d*f1

def f1poly(f1, l, pi, r):
    a, b, c, d = -l, l, -r-pi, pi
    return a*f1**3 + b*f1**2 + c*f1 + d
```
