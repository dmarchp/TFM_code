```python
# let's see the configuration:
i = 8
contactsPath = getConfigsPath() + '/contacts'
fname = getFilenameRoot(N, arena_r)
fnameSufix = getFilenameContactIntSufix(loops, interac_r)
df = pd.read_parquet(contactsPath + '/' + fname + f'_{str(i).zfill(3)}' + fnameSufix)
#cicles = pd.unique(df['cicleID'])
#gaps = cicles[1:] - cicles[:-1]
# gaps = gaps[gaps > 1]
df = df.query('cicleID == 7408') # remember cicle ID starts at 0
df = df.drop(columns=['cicleID'])
g = ig.Graph.DataFrame(df, directed=False)
dfpos = pd.read_parquet(getConfigsPath() + f'/PRW_nBots_{N}_ar_{arena_r}_speed_9_speedVar_2_{str(i).zfill(3)}.parquet')
```

```python
# aprox, els ticks correctes haurien de ser aquests, si no prova de fer +-1:
ticks = (517+7408)*3
# 517 seria el cicleID corresponent al primer tick de la trajectoria (havent descartat temps inicial), pero al generar els fitxers
# faig que el cicle Id corri sempre desde 0, per recuperar els ticks correctes ho he de sumar al cicleID abans de multiplicar pel "deltaticks = 3"
dfpos = dfpos.query('ticks == @ticks')
vertices_ids = [v['name'] for v in g.vs]
true_layout = []
for ide in vertices_ids:
    x, y = float(dfpos.query('ID == @ide')['x_position']), float(dfpos.query('ID == @ide')['y_position'])
    true_layout.append((x, y))
```

```python
true_layout = ig.Layout(true_layout)
```

```python
# les posicions no es corresponen amb la llista de contactes (?)
fig, ax = plt.subplots(figsize=(5,5))
ig.plot(
    g,
    target=ax,
    layout=true_layout,
    vertex_size=5,
    vertex_color="steelblue",
    vertex_frame_width=2.0,
    vertex_frame_color="white"
)
for i in range(N):
    coords = true_layout[i]
    circle1 = plt.Circle(tuple(coords), 1.5*10, color='xkcd:purple', alpha=0.5) # , clip_on=False
    ax.add_patch(circle1)
    #circle2 = plt.Circle(tuple(coords), interac_r*10, fill=False, edgecolor='xkcd:grey', ls='--')
    #ax.add_patch(circle2)
circle = plt.Circle(true_layout[-1], interac_r*10, fill=False, edgecolor='xkcd:grey', ls='--')
ax.add_patch(circle)
```
