```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import trackpy as tp
import sys
sys.path.append('../../')
```

```python
t1 = pd.read_csv('located_frames_Video1.csv')
t1=tp.link(t1,200, memory=400)
print('Before filtering:', t1['particle'].nunique())
t1f = tp.filter_stubs(t1, 300)
print('After filtering:', t1f['particle'].nunique())
```

```python
t1f_p0 = t1f.query('particle == 0')
print(max(t1f_p0.frame))
print(max(t1f.frame))
```

```python
pIDs = pd.unique(t1f['particle'])
for pID in pIDs:
    t1f_pID = t1f.query('particle == @pID')
    print(pID, min(t1f_pID.frame), max(t1f_pID.frame))

# are the they the same particle? 0,17,18 N10 lam03 video 1
```

```python
# some trial for n10 lam03 video 1
fig, ax = plt.subplots(1,1, figsize=(6,6))

t = t1f.query('particle == 0').iloc[-100:]
xmin, xmax = min(t.x), max(t.x)
ymin, ymax = min(t.y), max(t.y)
tp.plot_traj(t, colorby='frame', ax=ax, cmap='summer')


t = t1f.query('particle == 17').iloc[:100]
xmin = min(t.x) if xmin > min(t.x) else xmin
xmax = max(t.x) if xmax < min(t.x) else xmax
ymin = min(t.y) if ymin > min(t.y) else ymin
ymax = max(t.y) if ymax < max(t.y) else ymax
tp.plot_traj(t, colorby='frame', ax=ax)
ax.set(xlim=(xmin,xmax), ylim=(ymin,ymax))
```
