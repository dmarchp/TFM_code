```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
dpi=0.01
pi_lims = (0.01, 0.99)
dl=0.01
l_lims = (0.01,0.49)
Npis = int((pi_lims[1] - pi_lims[0])/dpi) + 1
Nls = int((l_lims[1] - l_lims[0])/dl) + 1
```

```python
x_pi = np.linspace(pi_lims[0], pi_lims[1], Npis)
```

```python
x_pi
```

```python
y_l = np.linspace(l_lims[0], l_lims[1], Nls)
```

```python
xv_pi, yv_l = np.meshgrid(x_pi, y_l, sparse=True)
```

```python
xv_pi
```

```python
#xgrid, ygrid = np.mgrid[pi_lims[0]:pi_lims[1]:complex(0,Npis), l_lims[0]:l_lims[1]:complex(0,Nls)]
xgrid, ygrid = np.mgrid[pi_lims[0]:pi_lims[1]:Npis, l_lims[0]:l_lims[1]:Nls]
```

```python
f2 = np.empty([3, Npis, Nls])
```

```python
f2.shape
```

```python
#for x in xv_pi[0]:
#    for y in yv_l[0]:
for i in range(Npis):
    x = xv_pi[0][i]
    for j in range(Nls):
        y = yv_l[j][0]
        #print(x,y)
        f2[i][j] = x+y #*random.random()
```

```python
f2.shape
```

```python
fig, ax = plt.subplots(figsize=(6,5))
X, Y = np.mgrid[pi_lims[0]:pi_lims[1]:complex(0,Npis), l_lims[0]:l_lims[1]:complex(0,Nls)]
im = ax.pcolormesh(X, Y, f2, shading='nearest')
fig.colorbar(im, ax=ax, aspect=25, shrink=0.75, pad=0.025)
```

```python
xgrid_pi, ygrid_l = np.mgrid[pi_lims[0]:pi_lims[1]:complex(0,Npis), l_lims[0]:l_lims[1]:complex(0,Nls)]
xgrid_pi, ygrid_l = np.around(xgrid_pi,2), np.around(ygrid_l,2)
fs = np.empty([3, Npis, Nls])
```

```python
for i,x in enumerate(xgrid_pi[:,0]):
    for j,y in enumerate(ygrid_l[0,:]):
        #print(i,j)
        fs[0,i,j] = x+y
        fs[1,i,j] = x-y
        fs[2,i,j] = x*y
```

```python
fig, ax = plt.subplots(figsize=(6,5))
im = ax.pcolormesh(xgrid_pi, ygrid_l, fs[0], shading='nearest')
fig.colorbar(im, ax=ax, aspect=25, shrink=0.75, pad=0.025)
```

```python
fig, ax = plt.subplots(1,3,figsize=(18,5))
im0 = ax[0].pcolormesh(xgrid_pi, ygrid_l, fs[0], shading='nearest')
im1 = ax[1].pcolormesh(xgrid_pi, ygrid_l, fs[1], shading='nearest')
im2 = ax[2].pcolormesh(xgrid_pi, ygrid_l, fs[1], shading='nearest')
fig.colorbar(im2, ax=ax[2], aspect=25, shrink=0.75, pad=0.025)
```
